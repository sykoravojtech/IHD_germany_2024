```python
# Define the softmax function
def softmax(col):
    e_x = np.exp(col - np.max(col))
    return e_x / e_x.sum()

def normalize(col):
    return (col - col.min()) / (col.max() - col.min())

# Apply softmax to a column
beds_df['softmax_Value'] = softmax(beds_df['Value'])

beds_df
```
